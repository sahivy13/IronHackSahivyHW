```python
# Import the choice function of the random module
# https://stackoverflow.com/questions/306400/how-to-randomly-select-an-item-from-a-list

import random

# Assign to a list the 3 possible options: 'stone', 'paper' or 'scissors'.

choices = ['stone', 'paper', 'scissors']

# Assign a variable to the maximum number of games: 1, 3, 5, etc ...

max_games = 10

# Assign a variable to the number of games a player must win to win.
# Preferably the value will be based on the number of maximum games

min_wins = 3

# Define a function that randomly returns one of the 3 options.
# This will correspond to the play of the machine. Totally random.

draw_computer = random.choice(choices)

# Define a function that asks your choice: 'stone', 'paper' or 'scissors'
# you should only allow one of the 3 options. This is defensive programming.
# If it is not stone, paper or scissors keep asking until it is.

if draw_computer != 'stone' or draw_computer != 'paper' or draw_computer != 'scissors':
    
    draw_computer = random.choice(choices)

# Define a function that resolves a combat.
# Returns 0 if there is a tie, 1 if the machine wins, 2 if the human player wins

draw_human = random.choice(choices)


def winner():
    if draw_human=='stone' and draw_computer=='stone':
        return 0
    elif draw_human=='stone' and draw_computer=='paper':
        return 1
    elif draw_human=='stone' and draw_computer=='scissors':
        return 2
    elif draw_human=='paper' and draw_computer=='paper':
        return 0 
    elif draw_human=='paper' and draw_computer=='scissors':
        return 1 
    elif draw_human=='paper' and draw_computer=='stone':
        return 2
    elif draw_human=='scissors' and draw_computer=='scissors':
        return 0
    elif draw_human=='scissors' and draw_computer=='stone':
        return 1
    elif draw_human=='scissors' and draw_computer=='paper':
        return 2 

    
# Define a function that shows the choice of each player and the state of the game
# This function should be used every time accumulated points are updated

def players_choice():
    return [draw_human,draw_computer]
    
# Create two variables that accumulate the wins of each participant

human_wins = 0
computer_wins = 0
num_games = 0

# Create a loop that iterates while no player reaches the minimum of wins
# necessary to win. Inside the loop solves the play of the
# machine and ask the player's. Compare them and update the value of the variables
# that accumulate the wins of each participant.

def choose_option():
    
    draw_human = input("Choose stone, paper or scissors: ")
    
    if draw_human in ["Stone","stone"]:
        draw_human = "stone"
    elif draw_human in ["Paper","paper"]:
        draw_human = "paper"
    elif draw_human in ["Scissors","scissors"]:
        draw_human = "scissors"
    else:
        print("I do not understand input, try again.")
    return draw_human

while num_games < max_games:

    draw_computer = random.choice(choices)
    draw_human = choose_option()
    print('The human and the computer chose the following choices respectively: ',players_choice())
    
    w = winner()
    
    if w == 0:
        num_games += 1
        print('This is a tie.')
        print('')
    elif w == 1:
        computer_wins += 1
        num_games += 1
        print('Computer wins the round.')
        print('')
        if computer_wins == min_wins:
            break
    elif w == 2:
        human_wins += 1
        num_games += 1
        print('Human wins the round.')
        print('')
        if human_wins == min_wins:
            break
    
# Print by console the winner of the game based on who has more accumulated wins
  
if human_wins > computer_wins:
    print('Human wins the game.')
elif human_wins < computer_wins:
    print('Computer wins the game.')
else:
    print('Max amount of games reached, and resulted in a tie.')
```
